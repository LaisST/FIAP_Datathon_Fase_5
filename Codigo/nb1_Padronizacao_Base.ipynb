## Padronização das bases de dados

As bases de dados PEDE2022, PEDE2023 e PEDE2024 apresentam inconsistências estruturais, tais como:

- divergência nos nomes das colunas;
- ausência de variáveis em determinados anos;
- segmentação dos dados em múltiplas abas.

Dessa forma, antes do início da análise exploratória, será realizada a etapa de padronização dos dadose e consolidação das bases em um único dataset.

Essa etapa tem como objetivo garantir a consistência, integridade e comparabilidade dos dados ao longo do tempo.

### Importar Bibliotecas

```python
# Tratativa da base
import pandas as pd
```

### Importar as Bases


```python
# Importar o arquivo do github

url = "https://raw.githubusercontent.com/LaisST/FIAP_Datathon_Fase_5/main/Bases/BASE%20DE%20DADOS%20PEDE%202024%20-%20DATATHON.xlsx"

# Importar cada aba em um df
df_2022 = pd.read_excel(url, sheet_name="PEDE2022")
df_2023 = pd.read_excel(url, sheet_name="PEDE2023")
df_2024 = pd.read_excel(url, sheet_name="PEDE2024")
```

### Renomear as colunas

Foi realizada uma análise prévia das colunas das bases para identificar diferenças de nomenclatura entre os anos.

```python
# Padronização de nomes de colunas
mapa_colunas = {
    "Matem": "Mat",
    "Portug": "Por",
    "Inglês": "Ing",
    "Defas": "Defasagem",
    "Fase ideal": "Fase Ideal",
    "Nome Anonimizado": "Nome",
    "Ano nasc": "Data de Nasc",
    "Idade 22": "Idade"
}

# Função para renomear com base nas colunas mapeadas
def padronizar(df):
    df = df.rename(columns=mapa_colunas)
    return df

df_2022 = padronizar(df_2022)
df_2023 = padronizar(df_2023)
df_2024 = padronizar(df_2024)
```

### Excluir Colunas

```python
# Exclusão das colunas não utilizadas
colunas_remover = [
    "Avaliador1", "Avaliador2", "Avaliador3",
    "Avaliador4", "Avaliador5", "Avaliador6",
    "Rec Av1", "Rec Av2", "Rec Av3", "Rec Av4"
]

def remover_colunas(df):
    return df.drop(columns=colunas_remover, errors="ignore")

df_2022 = remover_colunas(df_2022)
df_2023 = remover_colunas(df_2023)
df_2024 = remover_colunas(df_2024)
```

```python
# Exclusão das colunas duplicadas
def remover_duplicadas(df):
    return df.loc[:, ~df.columns.duplicated()]

df_2022 = remover_duplicadas(df_2022)
df_2023 = remover_duplicadas(df_2023)
df_2024 = remover_duplicadas(df_2024)
```

### Criar Colunas

```python
# Criar coluna IPP em 2022 (não existe)
if "IPP" not in df_2022.columns:
    df_2022["IPP"] = None
```

```python
# Criar coluna de ano
df_2022["Ano_Referencia"] = 2022
df_2023["Ano_Referencia"] = 2023
df_2024["Ano_Referencia"] = 2024
```

### Garantir as mesmas colunas para todas as bases

```python
# Garantir mesmas colunas em todas as bases
colunas_padrao = set(df_2022.columns) | set(df_2023.columns) | set(df_2024.columns)

def garantir_colunas(df):
    for col in colunas_padrao:
        if col not in df.columns:
            df[col] = None
    return df

df_2022 = garantir_colunas(df_2022)
df_2023 = garantir_colunas(df_2023)
df_2024 = garantir_colunas(df_2024)
```

### Juntar as Bases

```python
# Concatenar bases
df_final = pd.concat([df_2022, df_2023, df_2024], ignore_index=True)


# Salvar base tratada
df_final.to_csv("base_tratada.csv", index=False)

print("Base tratada com sucesso!")
print(df_final.shape)
```
